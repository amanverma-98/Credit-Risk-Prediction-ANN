```python
import pandas as pd

df = pd.read_csv("cs-training.csv")
df.head()
```

```python
#filling null values
df['MonthlyIncome'].fillna(df['MonthlyIncome'].mean(), inplace=True)
df['NumberOfDependents'].fillna(df['NumberOfDependents'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
#drop irrelevant columns
df.drop('Unnamed: 0', axis=1, inplace=True)
```

```python
df.describe()
```

```python
df['target'] = df['SeriousDlqin2yrs']
df.drop('SeriousDlqin2yrs', axis=1, inplace=True)
```

```python
#handling age
df = df[df['age'] >= 18]
```

```python
df.describe()
```

```python
#handling outliers as ANN is sensitive
cols = ['RevolvingUtilizationOfUnsecuredLines','DebtRatio','MonthlyIncome']

for col in cols:
    upper = df[col].quantile(0.99)
    df[col] = df[col].clip(upper=upper)
```

```python
df[df['target'] == 0].count()
#data is imbalanced
```

```python
X = df.drop('target', axis=1)
y = df['target']
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify = y, random_state=42) # stratify use kiya h kyuki dataset imbalanced h to distribution preserve krne ke liye
```

```python
# converting numpy to tensor
import torch
import numpy as np

X_train = torch.from_numpy(X_train.astype(np.float32))
X_val = torch.from_numpy(X_val.astype(np.float32))
y_train = torch.from_numpy(y_train.values.astype(np.float32))
y_val = torch.from_numpy(y_val.values.astype(np.float32))
```

```python
y_train = y_train.unsqueeze(1)  #changing size for use of bce loss
y_val = y_val.unsqueeze(1)
```

```python
#Dataset and DataLoader

from torch.utils.data import TensorDataset, DataLoader

train_data = TensorDataset(X_train, y_train)
val_data = TensorDataset(X_val, y_val)
train_loader = DataLoader(train_data, batch_size=64, shuffle=True)
val_loader = DataLoader(val_data, batch_size=64, shuffle=False)
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
import torch.nn as nn
class CreditRiskANN(nn.Module):
    def __init__(self, input_dim, hidden_dim, dropout):
        super().__init__()

        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            nn.BatchNorm1d(hidden_dim),
            nn.ReLU(),
            nn.Dropout(dropout),

            nn.Linear(hidden_dim, hidden_dim // 2),
            nn.BatchNorm1d(hidden_dim // 2),
            nn.ReLU(),
            nn.Dropout(dropout / 2),

            nn.Linear(hidden_dim // 2, 1)
        )

    def forward(self, x):
        return self.net(x)
```

```python
!pip install optuna
import optuna
from sklearn.metrics import roc_auc_score
```

```python
def objective(trial):

    #Hyperparameters
    hidden_dim = trial.suggest_int("hidden_dim", 32, 128, step=32)
    dropout = trial.suggest_float("dropout", 0.1, 0.5)
    lr = trial.suggest_float("lr", 1e-4, 1e-2,log = True)
    batch_size = trial.suggest_categorical("batch_size", [32, 64, 128])

    train_loader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    val_loader = DataLoader(val_data, batch_size=batch_size, shuffle=False)

    model = CreditRiskANN(
        input_dim=X_train.shape[1],
        hidden_dim=hidden_dim,
        dropout=dropout
    ).to(device)

    #Class imbalance handling
    pos_weight =((len(y_train) - y_train.sum()) / y_train.sum()).detach().clone().to(device)

    criterion = nn.BCEWithLogitsLoss(pos_weight=pos_weight)
    optimizer = torch.optim.Adam(model.parameters(), lr=lr)

    best_auc = 0
    patience = 3
    counter = 0

    for epoch in range(20):
        model.train()
        for xb, yb in train_loader:
            xb, yb = xb.to(device), yb.to(device)
            optimizer.zero_grad()
            loss = criterion(model(xb), yb)
            loss.backward()
            optimizer.step()

        #Validation
        model.eval()
        val_logits, val_labels = [], []

        with torch.no_grad():
            for xb, yb in val_loader:
                xb = xb.to(device)
                val_logits.append(model(xb).cpu())
                val_labels.append(yb)

        val_logits = torch.cat(val_logits)
        val_labels = torch.cat(val_labels)
        val_probs = torch.sigmoid(val_logits)
        val_auc = roc_auc_score(val_labels.numpy(), val_probs.numpy())

        #Early stopping
        if val_auc > best_auc:
            best_auc = val_auc
            counter = 0
        else:
            counter += 1
            if counter >= patience:
                break

    return best_auc
```

```python
#optuna
study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=10)

print("Best ROC-AUC:", study.best_value)
print("Best Parameters:", study.best_params)
```

```python
#final model training with best parameters
best = study.best_params

final_train_loader = DataLoader(
    train_data,
    batch_size=best["batch_size"],
    shuffle=True
)

final_val_loader = DataLoader(
    val_data,
    batch_size=best["batch_size"],
    shuffle=False
)

final_model = CreditRiskANN(
    input_dim=X_train.shape[1],
    hidden_dim=best["hidden_dim"],
    dropout=best["dropout"]
).to(device)

criterion = nn.BCEWithLogitsLoss(
    pos_weight=torch.tensor(
        (len(y_train) - y_train.sum()) / y_train.sum()
    ).to(device)
)

optimizer = torch.optim.Adam(final_model.parameters(), lr=best["lr"])
```

```python
#final model with early stopping
best_auc = 0
patience = 5
counter = 0

for epoch in range(50):
    final_model.train()
    for xb, yb in final_train_loader:
        xb, yb = xb.to(device), yb.to(device)
        optimizer.zero_grad()
        loss = criterion(final_model(xb), yb)
        loss.backward()
        optimizer.step()

    final_model.eval()
    logits, labels = [], []

    with torch.no_grad():
        for xb, yb in final_val_loader:
            xb = xb.to(device)
            logits.append(final_model(xb).cpu())
            labels.append(yb)

    logits = torch.cat(logits)
    labels = torch.cat(labels)
    probs = torch.sigmoid(logits)
    auc = roc_auc_score(labels.numpy(), probs.numpy())

    print(f"Epoch {epoch+1} | Val ROC-AUC: {auc:.4f}")

    if auc > best_auc:
        best_auc = auc
        counter = 0
    else:
        counter += 1
        if counter >= patience:
            print("Early stopping triggered")
            break
```
